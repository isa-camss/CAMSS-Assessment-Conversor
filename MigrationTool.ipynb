<h2>What does the notebook do?</h2><br>
                            
<div>This <b>Assessment Conversor</b> -a CAMSS Utilities notebook- is a module that receives RDF files in <a href="https://joinup.ec.europa.eu/collection/common-assessment-method-standards-and-specifications-camss/solution/camss-assessment-eif-scenario/release/v510">CAMSS EIF scenario 5.1.0</a> (CAMSS Solution former version) and transforms them into the latest version <a href="https://joinup.ec.europa.eu/collection/common-assessment-method-standards-and-specifications-camss/solution/camss-assessment-eif-scenario/release/v510">CAMSS EIF scenario 6.0.0</a> of the CAMSS EIF scenario.<br>


<h3>How do RDF files look like?</h3><br>
    RDF files are labelled distinctively depending on the CAMSS EIF scenario version, see examples below:<br><center>EIF-5.1.0-CAMSSAssessment_OASIS Open Data Protocol (Odata).ttl</center><br>        
    Assessments RDF file names are referred to by three digits ("EIF"), by the version of the assessment (usually three digits, "5.1.0"), and the naming of the specification, which optionally includes its version.<br>

<h3>And then?</h3><br>
    It is an automated conversor designed to convert CAMSS Assessment EIF Scenario files in RDF format.<br><br>
Once the CAMSS Assessments are processed, RDF files are stored locally in the user's machine in the 'out' folder (inside the project folder) as an RDF file containing all the assessments. This file it is said to contain the CAMSS Assessments graph, ready to be shared in CELLAR TripleStore. Additionally, in 'out', in their respective subfolders, individual RDF files of the individual converted assessments are also delivered.<br><br>
