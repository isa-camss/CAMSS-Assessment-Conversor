<h3>How do RDF files look like?</h3><br>
    RDF files are labelled distinctively depending on the CAMSS EIF scenario version, see examples below:<br>         <br><center>EIF-3.1.0-EIF-3.1.0-CAMSS Assessment ODLR 2.2_EIF Scenario_v1.0.0.ttl</center><br><center><center>EIF-3.0.0-EIF-3.0.0-CAMSS Assessment of XSD 1.1v.1.0.0.ttl</center><br></center><br><center>EIF-3.1.0-EIF-3.1.0-CAMSS Assessment - Standard for Public Code v1.ttl</center><br>        Assessments RDF file names are referred to by three digits, by the version of the assessment (usually three digits at the end of the file name), and the naming of the specification, which optionally includes its version.<br>

<h3>And then?</h3><br>
    It is an automated conversor designed to convert any CAMSS Assessment EIF Scenario files in RDF format.<br><br>
Once the CAMSS Assessments are processed, RDF files are stored locally in the user's machine in the 'out' folder (inside the project folder) as an RDF file containing all the assessments. This file it is said to contain the CAMSS Assessments graph, ready to be shared in CELLAR TripleStore. Additionally, in 'out', in their respective subfolders, individual RDF files of the individual converted assessments are also delivered.<br><br>
